## score by yandex

```python
import requests

def ya_correction(text):
    url = 'https://speller.yandex.net/services/spellservice.json/checkText'
    r = requests.post(url, data=f'text={text}'.encode(), headers={'Content-Type': 'application/x-www-form-urlencoded'})
    corrected = text
    try:
        for c in reversed(r.json()):
            if c.get('s'):
                corrected = corrected[:c['pos']] + c['s'][0] + corrected[c['pos'] + c['len']:] 
    except:
        print(text)
        raise
    return corrected
```

```python
ya_hypotheses = []
from tqdm import tqdm
for each_sent in tqdm(input_sentences):
    ya_hypotheses.append(ya_correction(each_sent))
```

```python
try:
    results = evaluate_spelling_corrector(input_sentences, 
                                          etalon_sentences, 
                                          ya_hypotheses)
    
    print("f1:")
    print(results['f_measure'])
    print("precision:")
    print(results['precision'])
    print("_____")
except Exception as e:
    print("Zero precision")
    pass
```

```python
# Print sentences with miscorrections for particular estimator
for num, out_sent in enumerate(ya_hypotheses):
    if out_sent != etalon_sentences[num]: 
        print("Input: %s" % input_sentences[num])
        print("Etaln: %s" % etalon_sentences[num])
        print("Hypot: %s" % out_sent)
#         print("Data Analysis Dict:" )
#         pprint.pprint(results_data[num])
    print("___")
```
